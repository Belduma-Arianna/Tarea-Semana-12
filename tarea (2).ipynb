```python
import numpy as np

# Crear una matriz 3D de ejemplo (Ciudades, Semanas, Días)
# Dimensiones: (Número de ciudades, Número de semanas, Número de días por semana)
# En este ejemplo: 2 ciudades, 3 semanas, 7 días a la semana
temperaturas = np.array([
    [  # Ciudad 1
        [20, 22, 21, 23, 24, 25, 23],  # Semana 1
        [21, 23, 22, 24, 25, 26, 24],  # Semana 2
        [22, 24, 23, 25, 26, 27, 25]   # Semana 3
    ],
    [  # Ciudad 2
        [15, 17, 16, 18, 19, 20, 18],  # Semana 1
        [16, 18, 17, 19, 20, 21, 19],  # Semana 2
        [17, 19, 18, 20, 21, 22, 20]   # Semana 3
    ]
])

# Nombres de las ciudades (opcional, para mejor visualización)
ciudades = ["Ciudad A", "Ciudad B"]

# Iterar a través de las ciudades y semanas para calcular el promedio
for i in range(temperaturas.shape[0]): # Iterar a través de las ciudades
    print(f"Promedios de temperatura para {ciudades[i]}:")
    for j in range(temperaturas.shape[1]): # Iterar a través de las semanas
        # temperatures.shape[1] gives the number of weeks
        promedio_semanal = np.mean(temperaturas[i, j, :])
        print(f"  Semana {j+1}: {promedio_semanal:.2f}°C")
    print("-" * 20) # Separador entre ciudades
```
